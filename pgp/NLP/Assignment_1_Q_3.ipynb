```python
'''
2. Write a program to enter a string from user and perform following tasks.
• Write a python function named “RemoveStopWords” which returns the string after removing stop words
• Count frequency of each stop word present in a string using the above function
• Plot a bar graph depicting stop words and their frequencies
'''
```

```python
def RemoveStopWords(user_input):
    stop_word = stopwords.words('english')
    inp =''
    stop_freq = {}
    if isinstance(user_input, str):
        wd = word_tokenize(user_input)
    elif isinstance(user_input, list):
        wd = user_input
    
    for w in wd:
        if w.lower() in stop_word:
            if w.lower() in stop_freq:
                stop_freq[w.lower()] = stop_freq[w.lower()] + 1
            else:
                stop_freq[w.lower()] = 1
        else:
            inp = inp + ' '+w
    
    return (inp,stop_freq)
    
    
wd = RemoveStopWords(user_input)

## Printing the sentence after removing the stop word
print(wd[0])

## Count of frequency of each Stop word
print('\nCount of frequency of each Stop word:')
for k,v in wd[1].items():
    print("%s: %s" % (k, v))
```

```python
## importing the library

from nltk.stem import WordNetLemmatizer
```

```python
## Function to Lemmatize the string

def Lemmatize(user_input):
    ## Tokenizing the word
    if isinstance(user_input, str):
        wd = word_tokenize(user_input)
    elif isinstance(user_input, list):
        wd = user_input
        
    lemmatizer=WordNetLemmatizer()
    
    word_lem={}

    for w in wd:
        if w.lower() in word_lem:
            pass
        else:
            word_lem[w.lower()] = lemmatizer.lemmatize(w.lower())
    
    return word_lem
```

```python
Lemmatize(user_input)
```

```python
## importing the library for Stemming

from nltk.stem import PorterStemmer
```

```python
## Function for Stemming - Porter Stemming

def Stemmed(user_input):
    ## Tokenizing the word
    if isinstance(user_input, str):
        wd = word_tokenize(user_input)
    elif isinstance(user_input, list):
        wd = user_input
    
    stemmer = PorterStemmer()
    
    word_stem={}

    for w in wd:
        if w.lower() in word_stem:
            pass
        else:
            word_stem[w.lower()] = stemmer.stem(w.lower())
    
    return word_stem
```
