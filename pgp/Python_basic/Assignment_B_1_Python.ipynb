```python
'''
EDU University wants to find out, whether class attendance by the students should be made compulsory.  For that, they gave you a dataset, SECTION1.csv  which includes the following attributes in the same order.
 

 
["Sno", "Department", "Branch", "Semester", "Course", "RegistrationNo", "StudentName", "Date", "Slot group", "Time", "ClassConducted", "ClassAttended"].

But after looking into the data, you reached to a conclusion that it is not in a proper format for analysing.  

You need to manipulate the data and create a dataframe with the following attributes:
      

[RegistrationNo, Name, Date1, Date2, Date3, …. DateN]
 

Each row corresponds to the data of a student. The column names of Date1, Date2, Date3.... DateN should be replaced with the unique dates from the dataset. Values under Date1, Date2 … columns should be the total number of classes attended by the student on that date.
 

Input Format
 

The first file to read will be SECTION1.csv, which contains the data as mentioned above. This file is in .csv format and is present at the location (res/SECTION1.csv)
The second file to read will be testcase.txt present at the location (res/testcase.txt) which contains the following data:
First line represents the number of testcase pairs.
Followed by a set of RegistrationNo and date pairs. Each element of a pair is on a separate line.
Output Format

For each pair in the testcase.txt, create an output file, outputn.txt  where n represents the pair number.
outputn.txt should be present at the location (output/outputn.txt) . This file should include RegistrationNo and the attendance value on that particular date.



e.g. testcase.txt contains the following data:

3
16fcs201114
08/12/18
16fcs201112
09/11/18
16fcs201140
18/12/18


output1.txt contains the following data:
16fcs201114
4

output2.txt contains the following data:
16fcs201112
2

output3.txt contains the following data:
16fcs201140
5


'''
```

```python
import pandas as pd
import numpy as np
import sys
sys.path.append('../')


def Attendance():
    sec_columns = ["Sno", "Department", "Branch", "Semester", "Course", "RegistrationNo", "StudentName", "Date", "Slot group", "Time", "ClassConducted", "ClassAttended"]
    df_sec = pd.read_csv('SECTION1.csv', sep=',', names=sec_columns)

    col_req = ['RegistrationNo', 'StudentName', 'Date', 'ClassAttended']
    df_sec_list = df_sec[col_req].copy()
    df_sec_list.rename({'StudentName': 'Name'}, inplace=True, axis=1)
    df_sec_list.set_index("Date", inplace=True)
    df_attendence = df_sec_list.groupby(["RegistrationNo", "Date"]).agg({"ClassAttended": ["sum"]})
    df_attendence.columns = ['ClassAttended']
    df_attendence = df_attendence.reset_index()
    # df_attendence.set_index("Date", inplace=True)

    # df_attendence.style.hide_index()

    try:
        fp = open('testcase.txt', 'r')

        while True:
            cnt = fp.readline().strip()
            if cnt.isdigit():
                cnt = int(cnt)
                break

        for i in range(cnt):
            reg = fp.readline().strip()
            dt = fp.readline().strip()
            cnt_att = df_attendence.loc[(df_attendence.RegistrationNo == reg) & (df_attendence.Date == dt) ].ClassAttended.to_string(index=False).strip()
            # print(len(cnt_att.strip()))
            out_file = "output"+str(i+1)+".txt"
            try:
                f = open(out_file, 'w')
                f.write(reg+"\n")
                f.write(cnt_att.strip())
            finally:
                f.close()

    finally:
        fp.close()

        
Attendance()
```

```python
import os
import pandas as pd
#quizfolder="res/quiz"
#markssheetlist=os.listdir(quizfolder)


def quiz(df_bt, file_name, cat_low, cat_up):

    df_qz = pd.read_csv('res/quiz/'+file_name)
    df_qz = df_qz[["Firstname", "Grade/10.00"]].copy()
    df_qz.rename({'Firstname': 'Name', 'Grade/10.00': 'Grade'}, inplace=True, axis=1)
    df_qz["Grade"] = pd.to_numeric(df_qz["Grade"], errors='coerce')
    df_qz["Name"] = df_qz["Name"].str.upper()
    #print(df_qz.dtypes)
    #print(df_qz.size)
    df_qz = df_qz.loc[(df_qz.Grade >= cat_low) & (df_qz.Grade < cat_up)]
    #print(df_qz)
    df_qz = pd.merge(df_bt, df_qz, on='Name', how='inner')
    return df_qz["Name"].count().astype(str)

def Categorize():

    try:
        fp = open('res/testcaseStudent.txt', 'r')

        while True:
            cnt = fp.readline().strip()
            if cnt.isdigit():
                cnt = int(cnt)
                break

        for i in range(cnt):
            batch = fp.readline().strip()
            check_cat = fp.readline().strip()
            cat_low = 0
            cat_up = 0

            if check_cat == "lessthan50":
                 cat_low,cat_up = 0,5
            elif check_cat == "between50and60":
                 cat_low,cat_up = 5,6
            elif check_cat == "between60and70":
                 cat_low,cat_up = 6,7
            elif check_cat == "between70and80":
                 cat_low,cat_up = 7,8
            elif check_cat == "greaterthan80":
                 cat_low,cat_up = 8,10.1

            #df_stud = pd.read_csv('res/studentlist.csv')
            #print(df_stud)

            df_bt = pd.read_csv('res/batchwiselist/'+batch)
            df_bt = df_bt[["studentName"]].copy()
            df_bt.rename({'studentName': 'Name'}, inplace=True, axis=1)
            df_bt["Name"] = df_bt["Name"].str.upper()
            #print(df_bt)

            out_file = "output/output"+str(i+1)+".txt"
            try:
                f = open(out_file, 'w')
                f.write(quiz(df_bt, 'quiz1.csv', cat_low, cat_up).strip()+"\n")
                f.write(quiz(df_bt, 'quiz2.csv', cat_low, cat_up).strip())
            finally:
                f.close()

    finally:
        fp.close()

    
Categorize()
```
